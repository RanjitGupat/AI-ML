```python
import pandas as pd
```

```python
var = pd.DataFrame({"Name" : ["a","c","d","b","a","c","d","b","a","c","d","b"],
                      "S_1" :[12,13,21,16,17,19,15,14,18,10,13,18],
                      "S_2" : [13,18,10,13,18,13,21,16,17,12,13,21]})
```

```python
var
```

```python
var_New = var.groupby("Name")
var_New
```

```python
for x,y in var_New:
  print(x)
  print(y)
  print()
```

```python
var_New.get_group("d")
```

```python
var_New.get_group("a")
```

```python
var_New.mean()
```

```python
var_New.max()
```

```python
var_New.median()
```
